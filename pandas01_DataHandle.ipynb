```python
##  평균이나 중앙값을 사용해서 변경
dirty_df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv.txt')
x =  dirty_df['Calories'].mean()
dirty_df['Calories'].fillna(x, inplace=True)
dirty_df
```

## 잘못된 포맷의 데이터

### 정상 포맷으로 변경

```python
## 22, 26행의 날짜데이터 정상화
dirty_df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv.txt')
dirty_df['Date'] = pd.to_datetime( dirty_df['Date'])
dirty_df
```

```python
## Date 열에 NaT 행 삭제 22행
dirty_df.dropna(subset=['Date'], inplace=True)
dirty_df
```

## 잘못된 데이터 처리

### 값을 변경

```python
## 7행 Duration 열 이상값, 450을 45로 변경
dirty_df.loc[7,'Duration'] = 45
dirty_df.loc[7]
```

```python
## big dataset 인 경우 규칙을 세워 변경 가능
for x in dirty_df.index:
    if dirty_df.loc[x, 'Duration'] >= 60:
        dirty_df.loc[x, 'Duration'] = 40
```

```python
dirty_df
```

```python
## 조건을 만족하는 행 삭제, 40보다 크면 삭제
for x in dirty_df.index:
    if dirty_df.loc[x, 'Duration'] > 40:
        dirty_df.drop(x, inplace=True)
dirty_df
```
