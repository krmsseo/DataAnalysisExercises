```python
no2 = air_quality[air_quality['parameter']=='no2'] ## filtering 
```

```python
no2.head(10)
```

```python
no2.sort_index().head()
```

```python
no2.sort_index().groupby('location')['value'].mean()
```

```python
no2_subset = no2.sort_index().groupby('location').head(2)
```

```python
no2_subset
```

```python
## pivot table 생성
no2_subset.pivot(columns='location', values='value')
```

```python
no2_subset.pivot(columns='location', values='value').plot()
```

```python
air_quality.pivot_table(
    values='value',
    columns='parameter',
    index='location',
    aggfunc='mean',
    margins=True)
```
