## 2. 난수 생성 함수

일반적인 난수 생성 뿐만 아니라 특정 분포를 따른 난수 생성 가능

머신 러닝에서 데이터셋으로 부터 표본 추출을 위한 특정분포 난수 배열 사용 가능

시뮬에이션에서 사용 가능

### 난수생성을 위한 seed 설정

난수 생성을 위한 seed 값을 설정 가능

서로 다른 난수열을 생성하기 위해서는 다른 seed값을 설정하여 난수 생성

```python
import numpy as np

np.random.seed(seed=123)
print(np.random.rand(10))
print(np.random.rand(10))
```

```python
np.random.seed(seed=123)
print(np.random.rand(10))
print(np.random.rand(10))
```

```python
print(np.random.randint(1,100, 50))
```

## 정규분포를 따르는 난수

normal(loc, scale, size)
- loc : 평균, 생략시 0
- scale : 표준편차, 생략시 1
- size : 표본수

```python
print(np.random.normal(size=100))
```

```python
nd = np.random.normal(size=100)
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(nd)
plt.show()
```

### 표준 정규분포를 따르는 난수 생성

randn(size)


### 이항분포를 따르는 난수 생성

binomial(n,p,size)

### 포아송 분포를 따르는 난수 생성

poisson(lam, size)

### t-분포를 따르는 난수 생성

standard_t(df, size)

### 카이제곱 분포를 따르는 난수 생성

chisquare(df, size)

```python
print(np.random.binomial(n=1, p=0.5, size=5))
print(np.random.binomial(n=1, p=0.5, size=5))
```

```python
print(np.random.poisson(lam=5, size=10))
print(np.random.poisson(lam=5, size=10))
```

```python
print(np.random.standard_t(df=5, size=10))
print(np.random.standard_t(df=5, size=10))
```

```python
print(np.random.chisquare(df=5, size=10))
print(np.random.chisquare(df=5, size=10))
```
